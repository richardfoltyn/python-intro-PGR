# Linear regressions with `lstsq()`

Consider the following quadratic relationship,
$$
y_i^* = a(x_i - b)^2 + c
$$
with parameters $a$, $b$ and $c$. Assume that we only observe it with additive measurement error,
$$
y_i = y_i^* + \epsilon_i = a(x_i - b)^2 + c + \epsilon_i, \qquad
\epsilon_i \stackrel{\text{iid}}{\sim} N\left(0, \sigma^2\right)
$$
Assume that the model is parametrised by
$a = 0.5$, $b = 1$, $c = 1$ and $\sigma=1.0$.

For a given sample, we can estimate the parameters by OLS by running the regression
$$
y_i = \beta_0 + \beta_1 x_i + \beta_2 x_i^2 + \epsilon_i
$$
*Note:* This does not estimate $a$, $b$ and $c$ directly, these are instead
going to be functions of $(\beta_0, \beta_1, \beta_2)$.

2. Create the true response variable $y_i^*$.
3. Draw 30 values for $\epsilon_i$ using the seed `123`.
4. Create the "observed" response variable $y_i$.

## Estimating coefficients using OLS

Recall that the OLS estimator $\mathbf{\beta}$ is implicitly defined by the linear equation system
$$
\mathbf{X}'\mathbf{X}\mathbf{\beta} = \mathbf{X}'\mathbf{y}
$$
While we tend to write the estimator as 
$$
\widehat{\mathbf{\beta}} = \left(\mathbf{X}'\mathbf{X}\right)^{-1} \mathbf{X}'\mathbf{y}
$$
we would usually not manually invert $\mathbf{X}'\mathbf{X}$ to obtain this solution directly. We instead use linear algebra routines 
to find the solution $\mathbf{x}$ for the linear system
$$
\mathbf{A}\mathbf{x} = \mathbf{b}
$$
where $\mathbf{A}$ is a coefficient matrix and $\mathbf{b}$ is vector. Clearly, in the case of OLS we have
$$
\begin{aligned}
    \mathbf{A} &= \mathbf{X}'\mathbf{X} \\
    \mathbf{b} &= \mathbf{X}'\mathbf{y} \\
    \mathbf{x} &= \mathbf{\beta}
\end{aligned}
$$

1. Create the matrix $\mathbf{X}'\mathbf{X}$ and the vector $\mathbf{X}'\mathbf{y}$.
2. Estimate the coefficient vector $\mathbf{\beta}$ by running NumPy's [`lstsq()`](https://numpy.org/doc/stable/reference/generated/numpy.linalg.lstsq.html):

    ```python
    beta, *rest = lstsq(XX, Xy, rcond=None)
    ```
3. Compute the predicted values $\widehat{y}_i = \mathbf{x}_i'\mathbf{\beta}$
4. Scatter-plot the raw data, and add lines for the true and the estimated relationships.
